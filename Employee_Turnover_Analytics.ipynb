Understanding Different Groups of Employees Who Left:
Clustering employees who have left the company can reveal distinct segments, such as:

Low Satisfaction and High Workload Group: This cluster consist of employees who had low job satisfaction but were assigned too many projects or had long working hours. These are typically high-risk employees for burnout.

High Performers Who Left: Employees who had high performance evaluations but left the company might suggest a lack of career growth opportunities or dissatisfaction despite doing well.

Underperformers or Low Engagement Group: Employees who consistently had poor performance reviews, low engagement, and eventually left the company. Understanding this group can help in identifying early signs of disengagement.

Perform 5-fold cross-validation model training and evaluate performance.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
data = load_iris()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
model = LogisticRegression(max_iter=200)
y_pred = cross_val_predict(model, X, y, cv=5)
report = classification_report(y, y_pred, target_names=data.target_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()
plt.figure(figsize=(8, 8))
sns.heatmap(report_df.iloc[:-1, :].T, annot=True, fmt='.2f', cmap='Blues', cbar=True)
plt.title('Classification Report')
plt.xlabel('Classes')
plt.ylabel('Metrics')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
y_pred = cross_val_predict(model, X, y, cv=5)
report = classification_report(y, y_pred, target_names=data.target_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()
plt.figure(figsize=(8, 8))
sns.heatmap(report_df.iloc[:-1, :].T, annot=True, fmt='.2f', cmap='Blues', cbar=True)
plt.title('Random Forest Classifier Classification Report')
plt.xlabel('Classes')
plt.ylabel('Metrics')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(n_estimators=100, random_state=42)
y_pred = cross_val_predict(model, X, y, cv=5)
report = classification_report(y, y_pred, target_names=data.target_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()
plt.figure(figsize=(8, 8))
sns.heatmap(report_df.iloc[:-1, :].T, annot=True, fmt='.2f', cmap='Blues', cbar=True)
plt.title('Gradient Boosting Classifier Classification Report')
plt.xlabel('Classes')
plt.ylabel('Metrics')
plt.show()
```

```python
y_bin = label_binarize(y, classes=[0, 1, 2])
n_classes = y_bin.shape[1]
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42)
clf = OneVsRestClassifier(RandomForestClassifier())
clf.fit(X_train, y_train)
y_score = clf.predict_proba(X_test)
tpr = dict()
roc_auc = dict()
fpr = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
plt.figure(figsize=(8, 8))
colors = ['aqua', 'darkorange', 'cornflowerblue']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'Class {i} (AUC = {roc_auc[i]:.2f})')
    plt.plot([0, 1], [0, 1], 'k--', lw=2)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve on RandomForestClassifier ')
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
turnover_probabilities = np.array([0.12, 0.45, 0.72, 0.85, 0.95])  
employee_data = pd.DataFrame({
    'Employee_ID': [f'E{i+1:03}' for i in range(len(turnover_probabilities))], 
    'Turnover_Probability': turnover_probabilities
})
def categorize_probabilities(prob):
    if prob < 0.20:
        return 'Safe Zone (Green)'
    elif 0.20 <= prob < 0.60:
        return 'Low-Risk Zone (Yellow)'
    elif 0.60 <= prob < 0.90:
        return 'Medium-Risk Zone (Orange)'
    else:
        return 'High-Risk Zone (Red)'
employee_data['Risk_Zone'] = employee_data['Turnover_Probability'].apply(categorize_probabilities)
print(employee_data[['Employee_ID', 'Turnover_Probability', 'Risk_Zone']])
```

1. Safe Zone (Green) (Score < 20%)
Characteristics: Employees in this category are considered low-risk for turnover.

Retention Strategies:

Maintain Engagement: Regularly check in with these employees to maintain their motivation and job satisfaction.
Recognition Programs: Implement recognition programs to celebrate their achievements, making them feel valued.
Career Development Opportunities: Provide opportunities for training and skill development to help them grow within the organization and encourage long-term commitment.

2. Low-Risk Zone (Yellow) (20% < Score < 60%)
Characteristics: Employees are somewhat at risk but not in immediate danger of leaving.

Retention Strategies:

Open Communication: Foster an environment where employees feel comfortable discussing any concerns or job dissatisfaction. Conduct regular one-on-one meetings to understand their needs.
Address Concerns: Identify and address any issues that may be contributing to their potential for turnover. This could include workload, job responsibilities, or team dynamics.
Professional Development: Offer opportunities for training and mentorship to help these employees advance in their careers, showing them that you are invested in their growth.

3. Medium-Risk Zone (Orange) (60% < Score < 90%)
Characteristics: These employees are at a higher risk of leaving and may have specific grievances or issues.

Retention Strategies:

Stay Interviews: Conduct stay interviews to understand their motivations and what would make them stay. Ask open-ended questions to uncover underlying issues.
Improve Work-Life Balance: Offer flexible work arrangements or additional time off to help these employees manage stress and improve their work-life balance.
Targeted Development Plans: Create personalized development plans based on their skills and aspirations, ensuring they see a clear path for advancement within the organization.
4. High-Risk Zone (Red) (Score > 90%)
Characteristics: These employees are at immediate risk of leaving and may have significant issues impacting their job satisfaction.

Retention Strategies:

Immediate Action: Implement urgent retention measures, such as personalized discussions to address their specific concerns.
Incentives: Offer tailored incentives, such as bonuses, salary increases, or additional benefits, to encourage them to stay.
Career Counseling: Provide career counseling sessions to discuss their future within the company, explore their goals, and align them with available opportunities.
Exit Interviews: If an employee decides to leave, conduct exit interviews to understand their reasons for leaving and use this feedback to improve the work environment for remaining employees.
Summary
Implementing these strategies can help improve employee retention across all zones. Regularly reassessing employee satisfaction and turnover risk can also ensure that strategies remain effective and relevant. By taking a proactive approach to engagement and retention, organizations can create a more positive work environment and reduce turnover rates.
